# <center> 2023-2024 Music Scores

In this notebook we will be analyzing if reading and language scores in 3rd and 4th grade have changed with the introduction of a music-based afterschool club into a student's curriculum.
 
Firstly, for this analysis we assume the groups are essentially even in skill. This could be a bit biased because students willing to participate in an after school program may have different factors affecting them such as a better support system at home/higher socioeconomic status or lower scores in general (hence their parents feeling the need to enroll them in an extracurricular activity). 

However, since this is less of a study and more of a data exploration, we'll assume that the control group and club groups are statistically equivalent prior to the school year. Further, improvement is a hard statistic to gauge because a student that begins at a 97% doesn't quite have the same opportunities for growth as a student beginning at 47%. 

### Step 1: Import Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
club_3 = pd.read_csv('3rd grade club.csv')
control_3 = pd.read_csv('3rd grade control.csv')
club_4 = pd.read_csv('4th grade club.csv')
control_4 = pd.read_csv('4th grade control.csv')
```

```python
c = ['Parent Letter Sent Home?', 'Parent Phone Call?']
club_3 = club_3.drop(columns = c)
club_3 = club_3.dropna()
control_3 = control_3.dropna()

club_4 = club_4.drop(columns = c)
club_4 = club_4.dropna()
control_4 = control_4.dropna()
```

```python
club_3_Rmean = club_3['Reading Growth'].mean()
club_4_Rmean = club_4['Reading Growth'].mean()
control_3_Rmean = control_3['Reading Growth'].mean()
control_4_Rmean = control_4['Reading Growth'].mean()
```

### Step 2: Initial Visualization

First we will visualize growth between the two groups

```python
r_means = [club_3_Rmean, control_3_Rmean, club_4_Rmean, control_4_Rmean]
columns = ["3rd Grade Club", "3rd Grade Control", "4th Grade Club", "4th Grade Control"]
```

```python
plt.bar(columns, r_means)
plt.xlabel('Groups')
plt.ylabel('Mean Reading Growth')
plt.title('Mean Reading Growth Comparison')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
club_3_Lmean = club_3['Language Growth'].mean()
club_4_Lmean = club_4['Language Growth'].mean()
control_3_Lmean = control_3['Language Growth'].mean()
control_4_Lmean = control_4['Language Growth'].mean()
l_means = [club_3_Lmean, control_3_Lmean, club_4_Lmean, control_4_Lmean]
```

```python
plt.bar(columns, l_means)
plt.xlabel('Groups')
plt.ylabel('Mean Language Growth')
plt.title('Mean Language Growth Comparison')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Step 3: Hypothesis Testing

As you can see above, it appears that in general the control groups are growing more than the club groups. However, let's do some hypothesis testing to see if the differences are significant. For this we will be using a student's T-test because we have a small n and unknown distribution. We will be testing for significance alpha of 0.05.

Our H0 is control mean = club mean, and our H1 is the opposite. 

```python
t_statistic_3r, p_value_3r = ttest_ind(club_3['Reading Growth'], control_3['Reading Growth'])

print("T-statistic For 3rd Grade Reading:", t_statistic_3r)
print("P-value for 3rd Grade Reading:", p_value_3r)
```

```python
t_statistic_3l, p_value_3l = ttest_ind(club_3['Language Growth'], control_3['Language Growth'])

print("T-statistic For 3rd Grade Language:", t_statistic_3l)
print("P-value for 3rd Grade Language:", p_value_3l)
```

```python
t_statistic_4r, p_value_4r = ttest_ind(club_4['Reading Growth'], control_4['Reading Growth'])

print("T-statistic For 3rd Grade Reading:", t_statistic_4r)
print("P-value for 3rd Grade Reading:", p_value_4r)
```

```python
t_statistic_4l, p_value_4l = ttest_ind(club_4['Language Growth'], control_4['Language Growth'])

print("T-statistic For 3rd Grade Language:", t_statistic_4l)
print("P-value for 3rd Grade Language:", p_value_4l)
```

### Step 4: Results

According to our hypothesis test, there is not enough evidence to indicate that there is a significant difference between the two groups.
